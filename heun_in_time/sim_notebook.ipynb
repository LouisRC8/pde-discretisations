### Imports Heun's Method Implementation 

```python
import numpy as np
from simulation import solve, plotsol
```

### Simulation Parameters

```python
# -------------------------
# Simulation Parameters
# -------------------------

# Space Discretisation
M = 100
L = 100
delta_x = L / M

# Time Discretisation
delta_t = 0.5
tmax = 100
N = int(tmax / delta_t)  # Number of time steps

# Courant Number
c = (np.sqrt(3)/2) - 1.0

# Advection Speed
a = c * (delta_x / delta_t)
```

### Plot Heavy Side Initial Solution, Solve, and Plot Final Solution

```python
# Initial Condition
u_n = np.ones(M)
u_n[: (M // 2)] = 0

# Plot Initial Solution
plotsol(u_n, 0, delta_t)
# Run Simulation
u_n = solve(u_n, c, N, delta_t)
plotsol(u_n, N, delta_t)
```

### Plot Sine Initial Solution, Solve, and Plot Final Solution

```python
# Initial Condition
x = np.linspace(-np.pi, np.pi, M)
u_n = np.sin(x)

# Plot Initial Solution
plotsol(u_n, 0, delta_t)
# Run Simulation
u_n = solve(u_n, c, N, delta_t)
plotsol(u_n, N, delta_t)
```
