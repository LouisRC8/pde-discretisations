### Imports Heun's Method Implementation 

```python
import numpy as np
from simulation import solve, plotsol
```

### Simulation Parameters

```python
# Simulation Parameters
delta_t = 0.5
M = 100
a = 1
L = 100
delta_x = L / M
N = int(L / delta_t)
c = (a * delta_x) / delta_t

# Initial Condition
u_n = np.ones(M)
u_n[: (M // 2)] = 0
```

### Plot Initial Solution, Solve, and Plot Final Solution

```python
# Plot Initial Solution
plotsol(u_n, 0)
# Run Simulation
u_n = solve(u_n, c, N)
plotsol(u_n, N)
```
